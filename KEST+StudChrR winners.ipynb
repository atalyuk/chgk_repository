```python
import pandas as pd
```

```python
studchr = pd.read_csv('medalists_studchr.csv')
kest = pd.read_csv('medalists_kest.csv')
```

```python
final = studchr.set_index('name').add(kest.set_index('name'), fill_value=0).reset_index()
cols = ['sum', 'chgk_sum', 'chgk_1.0', 'chgk_2.0', 'chgk_3.0', 'brain_sum', 'brain_1', 'brain_2', 'brain_3',  
        'ek_sum', 'ek_1', 'ek_2', 'ek_3', 'si_sum', 'si_1', 'si_2', 'si_3']
final = final.sort_values(ascending=False, axis=0, by=cols)
final = final.reset_index(drop=True)
final.head(10)
```
